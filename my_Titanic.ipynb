```python
raw_data = pd.read_csv('train.csv', index_col="PassengerId")
```

```python
clean_data = raw_data.drop('Cabin', axis=1)
clean_data
```

```python
median_age = clean_data["Age"].median()
clean_data['Age'] = clean_data['Age'].fillna(median_age)
```

```python
clean_data
```

```python
clean_data['Embarked'] = clean_data['Embarked'].fillna('U')
```

```python
preprocessed_data = clean_data
gender_colums = pd.get_dummies(preprocessed_data['Sex'], prefix='Sex')
embarked_colums = pd.get_dummies(preprocessed_data["Embarked"], prefix="Embarked")
```

```python

preprocessed_data = pd.concat([preprocessed_data, gender_colums], axis=1)
preprocessed_data = pd.concat([preprocessed_data, embarked_colums], axis=1)
```

```python
preprocessed_data
```

```python
pclass_colums = pd.get_dummies(preprocessed_data['Pclass'], prefix='Pclass')
preprocessed_data = pd.concat([preprocessed_data, pclass_colums], axis=1)
```

```python
preprocessed_data = preprocessed_data.drop(['Sex', 'Embarked', 'Pclass'], axis=1)
```

```python
preprocessed_data
```

```python
bins = [10, 20, 30, 40, 50, 60, 70, 80]
categorized_age = pd.cut(preprocessed_data['Age'], bins)
preprocessed_data['Categorized_age'] = categorized_age
preprocessed_data = preprocessed_data.drop(['Age'], axis=1)
```

```python
preprocessed_data
```

```python
preprocessed_data = preprocessed_data.drop(['Name', 'Ticket'], axis=1)
preprocessed_data
```

```python
preprocessed_data.columns
```

```python
preprocessed_data.index_col = 'PassengerId'
```

```python
preprocessed_data
```

```python
preprocessed_data = preprocessed_data.drop(['PassengerId'], axis=1)
```

```python
preprocessed_data
```

```python
preprocessed_data
```

```python
age_columns = pd.get_dummies(preprocessed_data['Categorized_age'], prefix='Age')
preprocessed_data = pd.concat([preprocessed_data, age_columns], axis=1)
preprocessed_data = preprocessed_data.drop(['Categorized_age'], axis=1)
```

```python
data = preprocessed_data
features = data.drop(['Survived'], axis=1)
labels = data['Survived']
```

```python
features
labels
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
features_train, features_validation_test, labels_train, labels_validation_test = train_test_split(features, labels, test_size=0.4)
features_validation, features_test, labels_validation, labels_test = train_test_split(features_validation_test, labels_validation_test, test_size=0.5)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(features_train, labels_train)
```

```python
lr_model.score(features_validation, labels_validation)
```
