```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.columns
```

```python
type(df)
```

```python
df.columns = df.columns.str.lower()
```

```python
df
```

```python
y = df['Survived']
data = df.drop(columns=['Survived'])
```

```python
data.info()
```

```python
data = data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
data.head()
```

```python
data.info()
```

```python
mean_age = data['Age'].mean()
data['Age'] = data['Age'].fillna(mean_age)
data.info()
```

```python
value_counts_series = data['Embarked'].value_counts()
index = (value_counts_series).argmax()
value_counts_series.index[index]
```

```python
type(value_counts_series.reset_index())
```

```python
data['Embarked']= data['Embarked'].fillna('S')
data.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
encoder.fit(data[['Sex', 'Embarked']])
```

```python
encoding_features = encoder.transform(data[['Sex', 'Embarked']])
```

```python
encoding_features[:, 0]
```

```python
data['Sex'] = encoding_features[:, 0].astype(int)
```

```python
data["Sex"]
```

```python
data = pd.get_dummies(data, columns=['Sex'])
```

```python
data = pd.get_dummies(data, columns=['Embarked'])
```

```python
data.info()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
lr = LogisticRegression()

cross_val_scores = cross_val_score(lr, data, y, cv=5, scoring='accuracy')
print("Ð—Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ Ð¼ÐµÑ‚Ñ€Ð¸ÐºÐ¸ accuracy Ð½Ð° ÐºÑ€Ð¾ÑÑ-Ð²Ð°Ð»Ð¸Ð´Ð°Ñ†Ð¸Ð¸: ", cross_val_scores)
print("Ð¡Ñ€ÐµÐ´Ð½ÐµÐµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð¼ÐµÑ‚Ñ€Ð¸ÐºÐ¸ accuracy Ð½Ð° ÐºÑ€Ð¾ÑÑ-Ð²Ð°Ð»Ð¸Ð´Ð°Ñ†Ð¸Ð¸: ", np.mean(cross_val_scores))
```

```python
knn = KNeighborsClassifier(n_neighbors=5)

cross_val_scores = cross_val_score(knn, data, y, cv=5, scoring='accuracy')
print("Ð—Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ Ð¼ÐµÑ‚Ñ€Ð¸ÐºÐ¸ accuracy Ð½Ð° ÐºÑ€Ð¾ÑÑ-Ð²Ð°Ð»Ð¸Ð´Ð°Ñ†Ð¸Ð¸: ", cross_val_scores)
print("Ð¡Ñ€ÐµÐ´Ð½ÐµÐµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð¼ÐµÑ‚Ñ€Ð¸ÐºÐ¸ accuracy Ð½Ð° ÐºÑ€Ð¾ÑÑ-Ð²Ð°Ð»Ð¸Ð´Ð°Ñ†Ð¸Ð¸: ", np.mean(cross_val_scores))
```

```python
lr.fit(data, y)
knn.fit(data, y)
```
